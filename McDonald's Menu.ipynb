## 4. Data Visulaization

### Creating a bar  chart based on the amount of fat in each category

```python
sns.barplot(x=df['Total Fat'], y=df['Category'])
```

### A histogram to show the distribution of Fat, Protein and Sugars in McDonalds menu

```python
sns.distplot(a=df['Total Fat'], label='Total Fat', kde=False)
sns.distplot(a=df['Protein'], label='Protein', kde=False)
sns.distplot(a=df['Sugars'], label='Sugars', kde=False)
plt.legend()
```

### A heatmap showing the correlation between variables with the lighter color showing higher correlation

```python
f, ax = plt.subplots(figsize=(18,15))
sns.heatmap(df.corr(), annot=True, linewidth=0.5, ax=ax)
plt.show()
```
